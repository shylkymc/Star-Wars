# Star Wars

The data set is obtained from FiveThirtyEight's GitHub [repository](https://github.com/fivethirtyeight/data/tree/master/star-wars-survey).
In this project I will clean and explore the data. The data has the answers about Star Wars Movie which are asked to Star Wars' fans.

## Opening and Exploring the Data

```python
import pandas as pd
star_wars= pd.read_csv("star_wars.csv", encoding="ISO-8859-1")
```

```python
star_wars.head(10)
```

```python
star_wars.info()
```

```python
star_wars.columns
```

There is a null entrie in "RespondentID" column. I will clean it.

```python
star_wars=star_wars[star_wars["RespondentID"].notnull()]
```

```python
print(star_wars["RespondentID"].isnull().head())
```

I will recheck the two column, I will clean null values and I will select the boolean values that are true or false.

```python
yes_no={
    "Yes": True,
    "No": False
}
```

```python
star_wars["Have you seen any of the 6 films in the Star Wars franchise?"]=star_wars["Have you seen any of the 6 films in the Star Wars franchise?"].map(yes_no)
```

```python
star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"]=star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"].map(yes_no)
```

```python
star_wars.head()
```

There are checkbook questions on some columns. It means that the series of movie is watched or not. I will change the columns as boolean values. I will change the column names.

```python

import numpy as np

movie_mapping = {
    "Star Wars: Episode I  The Phantom Menace": True,
    np.nan: False,
    "Star Wars: Episode II  Attack of the Clones": True,
    "Star Wars: Episode III  Revenge of the Sith": True,
    "Star Wars: Episode IV  A New Hope": True,
    "Star Wars: Episode V The Empire Strikes Back": True,
    "Star Wars: Episode VI Return of the Jedi": True
}

```

```python
for c in star_wars.columns[3:9]:
    star_wars[c]=star_wars[c].map(movie_mapping)
```

```python

star_wars = star_wars.rename(columns={
        "Which of the following Star Wars films have you seen? Please select all that apply.": "seen_1",
        "Unnamed: 4": "seen_2",
        "Unnamed: 5": "seen_3",
        "Unnamed: 6": "seen_4",
        "Unnamed: 7": "seen_5",
        "Unnamed: 8": "seen_6"
        })
```

```python
star_wars.head()
```

The following colunms have the rankings about the series. I will convert to columns to the numeric type and change the names.

```python
star_wars[star_wars.columns[9:15]]= star_wars[star_wars.columns[9:15]].astype(float)
```

```python
star_wars=star_wars.rename(columns={"Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.": "ranking_1",
        "Unnamed: 10": "ranking_2",
        "Unnamed: 11": "ranking_3",
        "Unnamed: 12": "ranking_4",
        "Unnamed: 13": "ranking_5",
        "Unnamed: 14": "ranking_6"
        })
```

```python
star_wars.head()
```

```python
print(star_wars.columns[9:15])
```

Let's visualize the mean of views.

```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
plt.bar(range(6), star_wars[star_wars.columns[9:15]].mean())
```

In previous steps, we tried to provide more understandable data. We changed some columns names and some value types.

In the last step, we calculated the mean of the rankings for each movie.

It can be seen from the bar plot that the last three movies are more popular.

```python
star_wars[star_wars.columns[3:9]].sum()
```

```python
plt.bar(range(6), star_wars[star_wars.columns[3:9]].sum())
```

We have seen before that the last three movies are most popular. In this step we can consider that the last 2 movies are watched more then others. The 4th movie is popular but it was not watched like others.  

Let's see the differences on views and rankings according to genders. 

```python
males= star_wars[star_wars["Gender"]=="Male"]
females=star_wars[star_wars["Gender"]=="Female"]
```

```python
plt.bar(range(6), males[males.columns[9:15]].mean())
plt.title("Male Viewers' Rankings")
```

```python
plt.bar(range(6), females[females.columns[9:15]].mean(), color="pink")
plt.title("Female Viewers' Rankings")
```

```python
plt.bar(range(6), males[males.columns[3:9]].sum())
plt.title('Male Viewers')
```

```python
plt.bar(range(6),females[females.columns[3:9]].sum(), color="pink")
plt.title('Female Viewers')
```

Female and male rankings have similir value for last three movie. However male's watching rate is highest then females.
